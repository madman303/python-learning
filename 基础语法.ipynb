**在位运算优先级中，优先级从高到低大体是：算术运算符 => 位运算符 => 关系运算符 => 逻辑运算符 => 赋值运算符**

```python
i = 0
while i*i < 10:
    i += 1
    if i == 3:
        break
    print(str(i) + '*' + str(i) + '=',i * i)
else:
    print('while over!')
```

```python
for item in range(10):
    if item == 3:
        break
    print(item)
else:
    print('For Over!')
```

```python
#break语句：用于强行推出循环体，不再执行循环体中剩余的语句
for item in range(10):
    if item == 3:
        break
        print(item)
```

```python
#continue语句：用于结束本次循环，跳过循环体中尚未执行的语句，接着进行终止条件的判断，以决定是否持续循环
for item in range(10):
    if item == 3:
        continue
    print(item)
```

```python
#计算水仙花数：通过while循环计算出水仙花数。
#提示：水仙花数是一个三位数，三位数各位的立方之和等于三位数本身
i = 100;r = 0;s = 0;t = 0

while i < 1000:
    r = i // 100
    s = (i - r * 100) // 10
    t = i - r * 100 - s * 10
    if i == (r ** 3 + s ** 3 +t ** 3):
        print("i = " + str(i))
        
    i += 1
```
